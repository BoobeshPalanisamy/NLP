```python
#Translate each word or sentence from English to Spanish, French and German 

import pandas as pd
df = pd.read_csv("English.csv")
df.describe()
```

```python
from translate import Translator
```

```python
from translate import Translator
English = (df["English words/sentences"].head(20)).tolist()
```

```python
German = []
Spanish = []
French = []
translator1= Translator(from_lang="english",to_lang="german")
translator2= Translator(from_lang="english",to_lang="spanish")
translator3= Translator(from_lang="english",to_lang="french")
for i in English:
  translation1 = translator1.translate(i)
  German.append(translation1)
  translation2 = translator2.translate(i)
  Spanish.append(translation2)
  translation3 = translator3.translate(i)
  French.append(translation3)
```

```python
df_trans = pd.DataFrame(list(zip(English,German,Spanish,French)),columns =['English','German','Spanish','French'])
df_trans
```

```python
#Create a program that needs to automatically correct that spelling from the word or a given sentence.
from autocorrect import Speller
from nltk.tokenize import word_tokenize
spell = Speller(lang ='en')
```

```python
def correct_spell(tokens):
  senten_corrected = ' '.join([spell(word) for word in tokens])
  return senten_corrected
```

```python
English = (df["English words/sentences"].head(100)).tolist()
for i in English:
  sente = word_tokenize(i)
  print(correct_spell(sente))
```
